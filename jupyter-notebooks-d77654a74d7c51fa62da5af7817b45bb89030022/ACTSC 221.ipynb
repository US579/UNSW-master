## Interest

Amount funciton $A(t)$ where $t \geq 0$ is the time.  
Amount we invest = **Principal**, $[A(0)]$ -> the amount at $t = 0$.

Suppose first period, we want the interest, then $I_{1} = A(1) - A(0)$.  
In generally, for the interest in the $n$-th period, it's $A_{n} = A(n) - A(n-1)$.

## Rates of Interests

**Defn.**  
The **effective rate of interest** is the amount of interest earned during period divided by the initial principal amount, assuming the interest is received at the end of the period.  
The rate at period $n$ is determined by
$$i_{n} = \frac{I_{n}}{A(n-1)} = \frac{A(n) - A(n-1)}{A(n-1)} = \frac{\text{Interest}}{\text{Amount at the Start}}$$

## Simple Interest

If the interest is earned at a _linear_ function of time, this is **simple interest**.

Interest earned after $t$ years is
$$I = P \times r \times t$$
where,

* $P$ is the initial principal
* $r$ is the _annual_ rate of simple interest
* $t$ is the time of in years

The accumulated value $A(t)$ of $P$ (also called the _future_ value, or the maturity value of $P$), is
$$A(t) = P + I = P \times (1 + r \times t)$$

So, the amount function of an initial principal $P$ invested for $t$ periods under a period interest rate of $r$ is given by
$$A(t) = P(1 + rt)$$

To get the present value, we have
$$A(0) = A(t)(1 + rt)^{-1}$$

### Example

Loan of \$10000. 5% simple interest.  
Then the amount function as a function of time $t$ is
\begin{align*}
A(t) &= P(1 + rt) \\
&= 10000(1 + 0.05t)
\end{align*}

## Day Count Convention

If time is given in actual number of days, there are two conventions to assess the actual interest charge, so called **exact interest** and **ordinary interest**.

In _exact interest_, we assume 365 days (this is the Canadian method), thus
$$I = Pr \times \frac{\text{total number of days}}{365}$$

In _ordinary interest_ (**Banker's Rule**), we assume 360 days in a year, thus
$$I = Pr \times \frac{\text{total number of days}}{360}$$

## Accumulated Values under Compound Interest

**Compound interest** is when interest can be earned on interest.

The formula for the amount function for an initial principal invested for $n$ years at compound interest $r$ is given by,
$$A(t) = P(1 + r)^{n}$$

## Present Values

The process of cash flowing back in time is **discounting** the values.

The idea that multiplying by $1+r$ carries values into the future, and dividing by $1+r$ carries values back to the present.

So, the present value ($PV$) of an amount $A(t)$ received in $n$ periods in the future is given by
$$A(0) = PV = \frac{A(t)}{(1+r)^{n}}$$

## Nominal Rates of Interest

So far, our examples of compound interest assume that the interest is received and reinvested at the end of each year.

In many cases, the actual frequency of interest payments may be more often than annually.  
For example, many bank accounts pay interest at the end of each month, so the interest received is reinvested monthly.

In such an instance, we would say that the interest on the account **compounds monthly**.

Notation $i^{(m)}$ to mean a **nominal**, or **stated annual rate** compounded $m$ times per year.  
Ex. $i^{(2)} = 10\%$ means the nominal rate is 10%, compound semi-annually.

In general, $\$1$ invested at $i^{(m)}$ will receive interest at a rate of $\frac{i^{(m)}}{m}$, $m$ times per year (received at the end of each $\frac{1}{m^{\text{th}}}$ of a year.)

Therefore, an initial principal, $P$, will accumulate to a final value, $A$, when invested at $i^{(m)}$ for $n$ periods to
$$A = P\left(1 + \frac{i^{(m)}}{m}\right)^{n}$$

### Equivalent Rates

Two rates are **equivalent** if a given amount of principal invested for the same length of time at either rate produces the same accumulated values.

Suppose we have $i^{(4)} = 10\%$, then to find the equivalent annual rate of interest $i^{(1)}$, we solve for
$$\left(1 + \frac{i^{(4)}}{4}\right)^{4} = \left(1 + \frac{10\%}{4}\right)^{4} = 1 + i^{(1)}$$
So we have $i^{(1)} = 10.381289\%$.

This annually compounded rate that is equivalent to the given nominal rate is called the **effective annual rate** (**EAR**).

## Varying Rates

Suppose $\$200$ in invested in an account that pays $i^{(1)}=5\%$, for the first 2 years, followed by $i^{(1)}=8\%$ for the next 3 years...  
By splitting this, $A(2)$ is computed by
$$A(2) = 200(1+0.05)^{2}$$
And that $A(5)$ is thus computed by
$$A(5) = A(2)(1+0.08)^{3}$$
Combing the two, we have
$$A(5) = 200(1+0.05)^{2}(1 + 0.08)^{3} =  \$277.77$$

### Equivalent Rates

As an aside, suppose instead this account paid an fixed annual rate of $i$.  
What is this rate such that the account accumulates to the same amount?

We solve for
$$200(1 + i)^{5} = 277.77 \Longrightarrow i = 6.789849\%$$

## Extra

As compounding frequency _increases_, equivalent rates _decrease_ because we will get more times that the interest will be included into subsequent interest calculation.

Also, EARs will increase as $m$ increases because we need to have the impact of point above included into the EAR (which is not part of the nominal rate).

## Dated Values

Suppose there are the choices of:

1. Receive $\$100$ in 1 year
1. Receive $\$110$ in 2 year

The amounts of money are not received at the same point in time.  
Need to know both how much money is being received and when the money is being received in order to determine the better option.

Consequently, all calculations involve **dated values**.

Suppose the interest rate is $5\%$, we can compute the dated value (or time value) of both options at a common point in time!

### Method 1 (Time 2)

Take the $\$100$ received at time 1 and reinvest it for one year at $5\%$, then at the end of year two it will have accumulated to
$$\$100(1 + 5\%)^{1} = \$ 105.00$$
So receiving $\$100$ in one year is equivalent to receiving $\$105$ in two years.

$\therefore$ option B is better

### Method 2 (Time 1)

To borrow an amount of money at time one that will accumulate to $\$100$ at time 2, so we discount the $\$110$ back one period, say $V_{1}$,
$$V_{1} = \frac{\$100}{(1 + 5\%)^{1}} = \$104.76$$

So, we see that option A is to receive $\$100$ in 1 year whereas option B is to receive $\$104.76$ in 1 year, so option B is better.

### Takeaway

The first method, the relevant values at time 2 (year 2) are calculated;
whereas method 2 we compute the relevant values at time 1 (year 1).  
The date we select to compute the values is often called the **focal point**, or **focal date**.

If a given option is better than another option to one focal point, it will be better at any focal point!

A common focal date to select is in fact now, or time 0.  
In this case, the time value is the _present value_.

### Method 3 (Time 0)

Calculating the present value of both options...  
We have
$$PV_{\$100} = \frac{\$100}{(1+5\%)^{1}} = \$95.24$$
and
$$PV_{\$110} = \frac{\$110}{(1+5\%)^{2}} = \$99.77$$

So once again, option B is better.

## Equivalent Amounts

We say that 2 different values $X$ and $Y$, where $Y$ is received $n$ periods later under a period interest rate of $i$ are **equivalent** if
$$Y = X(1 + i)^{n}~~~~~\text{or equivalently}~~~~~X = Y(1+i)^{-n}$$

## Unknown Rate

Recall,
\begin{align*}
FV = PV\left(1 + \frac{i^{(m)}}{m}\right)^{n}
\end{align*}
where $FV$ is future value, $PV$ is present value, $i^{(m)}$ is the annual interest rate compounded $m$ times per year, and $n$ is the number of periods the money is invested.

**Example**

Find $i^{(2)}$ such that $\$100$ will grow to $\$1000$ in 10 years.

\begin{align*}
1000 &= 100\left(1 + \frac{i^{(2)}}{2}\right)^{20} \\
\left(1 + \frac{i^{(2)}}{2}\right) &= \left(\frac{1000}{100}\right)^{\frac{1}{20}} = 1.122018 \\
i^{(2)} &\simeq 24.4\%
\end{align*}

## Unknown Time

**Example**

How long will it take for $\$100$ to grow to $\$1000$ if $i^{(4)} = 10\%$.

\begin{align*}
1000 &= 100\left(1 + \frac{10\%}{4}\right)^{n} \\
n \times \ln\left(1 + \frac{10\%}{4}\right) &= \ln\left(\frac{1000}{100}\right)
\end{align*}

Thus $n = 93.249958$ **periods**.  
Therefore, the answer in years is $T = \frac{n}{4} = 23.3124896$ years.

## Doubling Time

A common question is how long will it take for money to double at a given rate?  
If we assume we invest $\$1$, we need to determine the number of periods for this to grow to $\$2$.  
So we need to solve

$$2 = (1 + i)^{n} \Longrightarrow n = \frac{\ln 2}{\ln(1 + i)}$$

So it takes $\frac{\ln 2}{\ln(1 + i)}$ periods for money to double when invested at a period rate of $i$.

A common approximation for this is called the "**Rule of 72**", which states that doubling time $n$ is approximately

$$n \simeq \frac{72}{i}$$

Another approximation is "**Rule of 70**", which states that the doubling time $n$ is approximately

$$n \simeq \frac{70}{i}$$

**Remark**  
Rule of 72 seems to work better for interest rates _over_ $5\%$.  
Rule of 70 seems to work better for interest rates _under_ $5\%$.

## Working with Inflation

In many ways, problems in compound interest are just examples of general problems involving exponential growth.

**Inflation** is another example of exponential growth.

## Real Rate of Return

An interesting problem is to determine the future value of an investment, after we consider the affects of inflation.

> The **Real Rate of Return** is defined to be the growth in purchasing power available after we consider the effects of inflation.  
> This is distinct from the **nominal rate of interest**, which is the interest rate that does not adjust for inflation.
>
> When people speak of interest rates on a day-to-day basis, they are really talking about nominal rates.

We define **real rate of interest** as follows,
$$i_{real} = \frac{i - r}{1 + r}$$
where $i$ is nominal interest rate and $r$ is inflation rate.

## Taxes

Investors pay taxes on interest earned.

The **after tax interest rate** is calculated as follows,
$$i_{\text{after tax}} = i(1 - T)$$
where $i$ is the nominal interest rate and $T$ is the tax rate.

## Taxes and Inflation

With _real rate of interest_ and _after tax interest rate_, we can calculate a **real after tax interest rate** as follows,
$$i_{\text{real after tax}} = \frac{i(1 - T) - r}{1 + r}$$
where $i$ is nominal interest rate, $T$ is tax rate, and $r$ is inflation rate.

## Rates of Discount

Interest paid at the start of the loan is being paid **in advance**, as opposed to being paid at the end of the loan, which is being paid **in arrears**.

When interest is paid in advance, it is a rate of discount, denoted $d$.  
So, if the loan principal is $P$, the amount you will receive today is then
$$\text{Amt Received} = \text{Principal} - \text{Interest} = P - P \times d = P(1-d)$$

The **effective rate of discount over period** $n$, denoted $d_{n}$, is the ratio of the cost of the loan (or the amount of interest) to the amount at the _end_ of the year,
$$d_{n} = \frac{A(n) - A(n-1)}{A(n)}$$

Discount is paid at the _beginning_ of the year based on the balance at the _end_ of the year;
while interest is paid at the _end_ of the year, based on the balance at the _beginning_ of the year.

### Converting to Rates of Interest

\begin{align*}
d_{n} &= \frac{A(n) - A(n-1)}{A(n)} \\
&= \frac{P(1+i)^{n} - P(1+i)^{(n-1)}}{P(1+i)^{n}} \\
&= \frac{i}{1+i} \\
i &= \frac{d}{1 - d}
\end{align*}

### Present and Future Values

For rates of discount,
$$FV = \frac{PV}{(1-d)^{t}} \iff PV = FV(1-d)^{t}$$

Whereas for rates of interest,
$$FV = PV(1+i)^{t} \iff PV = \frac{FV}{(1+i)^{t}}$$

## T-Bills

Treasury Bills are short term notes issued by governments.  
The governments use T-bills as a way to borrow money from the public over a short time frame.

### Canadian T-Bills

In Canada, the price of a T-Bill can be computed by discounting the face value using **simple interest** and dividing the exact number of days by _365_.

**Example**  
Compute the price of a 91 day T-bill if the rate is 5%.  
Assume a face value of \$1,000.  
So we get
$$P = \frac{1000}{\left(1 + \frac{91}{365}5\%\right)} = 987.69$$

### US T-Bills

In the States, the price is computed by discounting the face value using **simple discount** and dividing the exact number of days by _360_.

**Example**  
Compute the price of a 91 day US T-bill if the rate is 5%.  
Assume a face value of \$1,000.  
So we get
$$P = 1000\left(1 - \frac{91}{360}5\%\right) = 987.36$$

## Accumulated Values of Annuities

### Simple Annuities

An **annuity** is a periodic payment of periodic withdrawal.

An **ordinary** annuity is that the payments are due at the end of each period.  
A **simple** annuity is that the interest compounding period corresponds to the payment frequency.

The **accumulation factor**, "s n angle i", is
$$s_{n\rceil i} = \frac{(1+i)^{n} - 1}{i}$$
where $n$ is the total _number_ of payments and $i$ is interest rate _per payment period_.

So the future value is calculated by
$$S = R \times s_{n\rceil i}$$
where $R$ is the payment amount.

## Present Value of Annuities and Loans

### Present Value of Annuities

We discount the cash flows in an annuity in order to compute the present value of the payments.

The **discount factor**, "a n angle i", is
$$a_{n\rceil i} = \frac{1 - (1+i)^{-n}}{i}$$
where $n$ is the total _number_ of payments and $i$ is interest rate _per payment period_.

So the present value is calculated by,
$$PV = R \times a_{n\rceil i}$$
where $R$ is the payment amount.

### Loans

A major category of problems where we will use the discount factor is in the calculation of _loan payments_.

**Example**  
Loan of \$30,000, monthly payment of 5 years, interest rate of $i^{(12)} = 4\%$.
\begin{align*}
30000 &= R \times a_{\overline{12 \times 5}| \frac{4\%}{12}} \\
R &= 552.50
\end{align*}

## Calculating the Number of Payments

Suppose we make a loan of $PV$, with $R$ and $i$, then need to find $n$,
$$PV = R \times a_{n \rceil i} = R \times \left(\frac{1-(1+i)^{-n}}{i}\right)$$

**Example**  
Loan of \$20,000 and monthly payments of \$500. How many payments will I have to make if $i^{(12)}=12\%$?

\begin{align*}
20000 &= 500 \times a_{n \rceil \frac{12\%}{12}} \\
&= 500\left(\frac{1 - (1 + 1\%)^{-n}}{1\%}\right) \\
&\vdots \\
n &= -\frac{\ln 0.6}{\ln(1 + 1\%)} \simeq 51.3~\text{months}
\end{align*}

Need to make a bit over 51 payments... 52 payments is too much.  
There are 2 options:

1. The first is to make a slightly larger final payment in order to fully reply the loan (called the **balloon payment**). 
2. The second (and more common) is to make a smaller final concluding payment one period later (called the **drop payment**).

**Balloon payment**  
Present value can be seen as the sum of the present value of the 51 regular payments, plus the present value of the additional "top up" payment at time 51.  
This gives,
$$20000 = 5000\times a_{51\rceil 1\%} + X(1+1\%)^{-51}$$
Solving gives $X = \$167.66$.  
So the final balloon payment is $\$500 + \$167.66 = \$667.66$.

**Drop payment**  
If making a payment at time 52, then we have,
$$20000 = 5000\times a_{51\rceil 1\%} + X(1+1\%)^{-52}$$
Solving gives $X = \$169.33$.

## Different Focal Dates

### Deferred Annuities

A **deferred annuity** is an annuity with a delayed first payment by $m$ periods, i.e. the first payment occurs on $(m+1)$-th period.

The present value is then,
$$P.V. = \frac{a_{n\rceil i}}{(1+i)^{m}} = a_{n\rceil i}\times(1 + i)^{-m}$$

Accumulated value is given by,
$$A.V. = s_{n \rceil i}$$

**Example**  
Sally wishes to purchase an annuity for her retirement.  
Sally is 30 years old today, and she wants the first payment to occur in 35 years when she turns 65.  
She would like the annuity to make 20 equal annual installments of $\$20,000$.  
How much will the annuity cost today, if $i^{(1)}=10$.

$$P.V. = \$20,000 \times (1 + 10\%)^{(30 - 64)}\times a_{20\rceil 10\%} = \$6,664.85$$

### Annuities Due

An **annuity due** is an annuity where the payments are made at the _start_ of each period.  
$\ddot{a}_{n\rceil i}$ is equivalent to present value of a regular annuity of $n - 1$ payments, plus the payment at time 0, so
$$\ddot{a}_{n\rceil i} = a_{\overline{n-1}\rceil i} + 1$$

Then the accumulated value $\ddot{s}_{n\rceil i}$ is given by,
$$\ddot{s}_{n\rceil i} = s_{n\rceil i}(1 + i)$$

### All in all...

$$(1 + i)a_{n\rceil i} = \ddot{a}_{n\rceil i}$$
$$(1 + i)^{n}a_{n\rceil i} = s_{n\rceil i}$$

## General Annuities

Simple annuities: equal payments at intervals which _correspond_ to the compound frequency of the interest rates.

General annuities: payment frequency and compound frequency _not_ equal.

One solution: find equivalent interest rate.

## Perpetuities

**Perpetuities** are annuities that continuing paying forever.

### Simple Perpetuities

For present value, then
$$a_{\infty\rceil i} = \lim_{n\to\infty}a_{n\rceil i} = \frac{1}{i}$$

**Example**

Find the present value of a perpetuity that pays $10$ at the end of each year, assuming $i^{(1)} = 10\%$.
$$P.V. = \$10 \times a_{\infty \rceil 10\%} = \$10 \times \frac{1}{10\%} = \$100$$

### General Perpetuities

Payment frequency and compound frequency _not_ equal.  
Thus, find equivalent interest rate and continue.

**Example**

What is the present value of a perpetuity that pays $250$ per month, with the first payment 2 years from today?   
Assume $i^{(1)}=11\%$.

We have then,
$$\left(1 + 11\%\right) = \left(1 + \frac{i^{(12)}}{12}\right)^{12} \Longrightarrow i^{(12)} = 10.4815\%$$
So,
$$P.V. = \$250 \times a_{\infty \rceil \frac{i^{(12)}}{12}} \times \left(1 + \frac{i^{(12)}}{12}\right)^{-23}$$
We use 23 period because that's one period before first cash flow.

## Increasing Annuities

### Geometric Growth

Payments are growing at a rate of $g$.  
Common application is when the growth rate is the rate of inflation.

Finding the present value,
$$P.V. = R \times \frac{1 - \left(\frac{1+g}{1+i}\right)^{n}}{i - g}~~~~~\text{where } i \neq g$$

#### Geometric Growth Perpetuities

$$P.V. = \frac{R}{i - g}$$

Perpetuities due:

$$P.V. = \frac{R}{i - g}(1 + i)$$

### Arithmetic Growth

Payments are growing by increments of $P$.

Finding the present value, denoted $(Ia)_{n\rceil i}$, of the **increasing annuity**,
$$(Ia)_{n\rceil i} = \frac{\ddot{a}_{n\rceil i} - n(1+i)^{-n}}{i}$$

So, combining, we have
$$P.V. = R \times a_{n\rceil i} + P \times (Ia)_{n-1\rceil i} \times (1 + i)^{-1}$$

#### Arithmetic Growth Perpetuities

$$P.V. = \left(\frac{R}{i} + \frac{P}{i^{2}}\right)$$

Perpetuities due:

$$P.V. = \left(\frac{R}{i} + \frac{P}{i^{2}}\right)(1 + i)$$

### Accumulated Values

We let $(Is)_{n\rceil i}$ denote the accumulated value of an annuity on the date of the last payment.
$$(IS)_{n\rceil i} = \frac{s_{n+1\rceil i} - (n+1)}{i}$$
So, combining, we have
$$C.V. = R \times s_{n\rceil i} + P \times (Is)_{n-1\rceil i}$$

OR

$$C.V. = (R-P) \times s_{n\rceil i} + P \times (Is)_{n\rceil i}$$

## Repayment of Debts

### Computing Debt Payments

When making payments on a loan, the payments are essentially of 2 parts:

1. paying the interest charges on the loan
2. reducing the principal

In this case was say that the principal balance is being **amortized** over the life of the loan.

**Example**

Suppose John borrows $100,000$ at an interest rate of $i^{(12)}=12\%$, and he pays off the loan over 10 years.  
What are the monthly payments?

Solving,
$$\$100,000 = R \times a_{120\rceil 1\%} \Longrightarrow R \simeq \$1,434.709484$$
Therefore, we typically round the payment up, say to the nearest cent.  
Thus, John actually pays $1434. 71$.

In this case, John is slightly overpaying on the loan.  
Consequently, the final payment will be lower than the regular payment.  
So, John will make only 119 regular payments of $1,434.71$ and one final payment that is lower.

So, we solve for
$$\$100,000 = \$1,434.71 \times a_{119\rceil 1\%} + X \times (1 + 1\%)^{-120}$$
which yields final payment, $X = 1,434.59$ (rounded to nearest cent).

### Amortization Schedules

The way we record the interest charges, the receipt of payments and the adjusted loan balance is to construct an **amortization schedule**.  
The amortization schedule is a table of values showing the remaining balance on the loan and records the new interest charges and payments made on the loan.

How was this schedule built? Here are the key points:

* each row corresponds to a payment period on the loan
* opening balance of one row is the closing balance of the previous row
* interest charge on the loan during a period is simply the opening balance times the interest rate
* payments are fixed, except for potentially lower last payment
* closing balance is the opening balance plus the interest charged less the payments made

### General Amortization Schedules

Given a general amortization schedule (and by **general** we mean without doing any rounding of fractional cents), we can actually determine the values in each row without needing to construct the entire table.

Let $A$ be the initial principal balance, and let $R$ be the periodic payment, assuming $n$ payments under an interest rate $i$, then

$$A = Ra_{n\rceil i}$$

Let $A_{k}$ be the remaining principal balance after $k$ payments are made, then

$$A_{k} = Ra_{n-k\rceil i}$$

The interest component paid as part of the $k^{th}$ payment, $I_{k}$, is

\begin{align*}
I_{k} &= i \times A_{k-1} \\
&= R \times (1 - (1+i)^{-(n-k+1)})
\end{align*}

The principal payment paid as part of the $k^{th}$ payment, $P_{k}$ is

\begin{align*}
P_{k} &= R - I_{k} \\
&= R \times (1+i)^{-(n-k+1)}
\end{align*}

## Debt Refinancing

When calculating a balance on a loan at some point in the middle of a loan, we can either look backwards at the payments already made, or we can look forwards to the payments yet to come.

**Example**

Borrowed $20,000$ at an interest rate of $i^{(12)}=6\%$ amortized over 5 years, paid monthly.  
After making 20 payments on time, misses _5 consecutive_ payments.  
The bank agreed to allows restart of making payments, but they would not agree to extend the original amortization period (thus the loan must be fully paid under the new terms on the same date it would have been under the original terms).  
What was the original monthly payment, and what will it be under the new terms?

### Retrospective Method

> Looking at the past payments in order to calculate the outstanding balance

**Example: Finding original payment**:

\begin{align*}
20000 &= R \times a_{5\times 12\rceil \frac{6\%}{12}} \\
R &= 386.6560
\end{align*}

Rounding up gives $R = 386.66$.

**Example: Finding new payment**:

To calculate the new terms, we need to know the outstanding balance when the loan is renegotiated, which will be after 25 months.  
The value of the new payments is denoted by $R^{*}$.

The balance at time 20, $B_{20}$, is the accumulated value of the loan, less the accumulated value of the payments,

\begin{align*}
B_{20} &= 20000 \left( 1 + \frac{6\%}{12} \right)^{20} - 386.66 \times s_{20\rceil\frac{6\%}{12}} \\
&= 13986.13
\end{align*}

Carrying this forward 5 periods yields the balance at time 25, $B_{25}$,

\begin{align*}
B_{25} &= B_{20} \left( 1 + \frac{6\%}{12} \right)^{5} \\
&= 13986.13 \times \left( 1 + \frac{6\%}{12} \right)^{5} \\
&= 14339.29
\end{align*}

So the new payment is,

\begin{align*}
14339.29 &= R^{*} \times a_{35\rceil\frac{6\%}{12}} \\
R^{*} &= 447.61
\end{align*}

### Prospective Method

> Looking at the normal payments to come in the future in order to find the outstanding balance

So $B_{20}$ must equal the present value of the future loan payments.  
Note that there are 39 more regular payments of $386.66$ plus the one final payment of $386.38$ (left as an exercise) in 40 periods.  
Therefore,

\begin{align*}
B_{20} &= 386.66 \times a_{39\rceil \frac{6\%}{12}} + 386.38 \left( 1 + \frac{6\%}{12} \right)^{-40} \\
&= 13986.13
\end{align*}

Rest is the same as before.

## Mortgages

In Canada, mortgage payments are typically made _monthly_, but the interest rate the banks quote on mortgages is often a _semi-annual rate_.  
Thus, we need to convert the interest rate before we do the final computations.

> The **amortization** (or **amortization period**) is the length of time it would take to pay off the entire mortgage loan assuming no change in interest rates and no missed payments.

### Term of a Mortgages

> The **term** of a mortgage is the length of time that the terms and conditions of the mortgage have legal effect.
> 
> In most instances, the significance of this is that the interest rate is essentially fixed for the term of the mortgage.  
> At the end of the term, the remaining mortgage balance must be repaid in full, or refinanced (typically at a new rate).

**Example**  
$200,000$ mortgage at an interest rate of $i^{(2)}=6\%$.  
Assume the mortgage has a 5 year term and a 25 year amortization.  
Assume that after the term expires, interest rates increase to $i^{(2)}=10\%$.  
What will the new monthly payments be?

Initial monthly payment of $1279.61$.  
The balance at time 60, $B_{60}$ is given by

\begin{align*}
B_{60} &= 200000 \times \left( 1 + \frac{i^{12}}{12} \right)^{60} - 1279.61 \times s_{60\rceil \frac{i^{(12)}}{12}} \\
&= 268783.28 - 89109.64 \\
&= 179673.63
\end{align*}

Now that we have the balance, we can compute the new payments.  
Since we have been paying the mortgage for 5 years, we would expect the new amortization period to now be only 20 years, instead of 25.  
Thus,

$$179673.63 = PMT \times a_{20\times 12 \rceil \frac{i^{(12)}}{12}}$$

Also, new interest rate, so

$$\left( 1 + \frac{10\%}{2} \right)^{2} = \left( 1 + \frac{i^{(12)}}{12} \right)^{12} \Rightarrow i^{(12)} = 9.7978\%$$

Hence the new payment is

\begin{align*}
PMT &= \frac{179673.63}{a_{20\times 12 \rceil \frac{i^{(12)}}{12}}} \\
&= 1709.89
\end{align*}

## Sinking Funds

In many instances, it is prudent to place funds aside in order to satisfy some future expected obligation, such as paying the principal amount when a bond comes due, or in order to purchase an asset.

Such a fund is called a **sinking fund**.

**Example**  
The board of homeowners in a townhouse complex determined that the parking lot will need to be rebuilt in 10 years.  
If the lot were rebuilt now, it would cost $100,000$.  
In order to fund this expense, the board has decided to make monthly deposits into a sinking fund.  
Assume the deposits are made into an account earning $i^{(12)}=3\%$.  
Assume also that inflation over the next 10 years is expected to be $1\%$ per year.  
How much should the deposits be?

\begin{align*}
R \times s_{12\times 10 \rceil \frac{3\%}{12}} &= 100,000 \times (1 + 1\%)^{10} \\
R &= 790.48
\end{align*}

## Bond Basics

### Introduction to Bonds

#### Bond Features

Bonds make up a large part of all government borrowings.  
By purchasing a bond, you are lending money to the issuer.

A major difference between a bond and a simple loan is that a bond is typically **marketable**, meaning the initial purchaser of the bond is able to sell the bond _prior_ to the maturity of the loan.  
Bonds are traded between investors in the **bond markets** which are some of the largest and most important markets in finance.

The bond contract (also called the **bond indenture**) outlines the terms of the loan.  
The terms include:

* **face value**
    * par value of the note
    * typically the principal amount that is paid when the bond matures
* **maturity date**
    * date when the bond matures, and the face value is paid to the investor
    * varies from 30 days to 30 years or more from _date of issue_
* **coupon rate**
    * percentage of the face value paid annually, essentially as interest on the bond
* **coupon frequency**
    * frequency of coupon payment
    * most common frequency is semi-annually
* **embedded options**
    * only some bonds allow this
    * e.g., issuer to repurchase the bond early (essentially this is an early repayment option); or options that allow the holder to redeem the bond early (essentially an early withdrawal option)

A simple bond with a fixed coupon rate, a fixed maturity date and no embedded options are sometime called **bullet bonds**.

#### Bond Pricing

Timeline of cash flow from a bond paying a coupon $C$ with a face value of $F$ maturing in $n$ periods.

```
          C         C         C          C+F
|---------|---------|---------| o o o o o |
0         1         2         3           n
```

Present value:
$$\text{Value} = C \times a_{n\rceil i} + \frac{F}{(1+i)^{n}}$$

\begin{definition}\label{def:YieldToMaturity}
The (annualized) rate that we use to discount the cash flows of the bond is called the **Yield to Maturity** (YTM) of the bond.  
Since bonds usually pay semi-annually, the YTM is quoted with semi-annual compounding.
\end{definition}

Therefore, the formula for the value of a coupon bond paying semi-annual coupons is
$$\text{Value} = C \times a_{n\rceil \frac{YTM}{2}} + \frac{F}{\left(1+\frac{YTM}{2}\right)^{n}}$$
where $YTM$ is the yield to maturity of the bond.

\begin{example}\label{example:YTMExample}
Consider a $5\%$ coupon Government of Canada 10 year bond with a YTM of $6\%$.  
Find the value of this bond.

**Solution**  
Since Government of Canada bonds pay coupons semi-annually, each coupon payment (based on a 1000 face value) is given by
$$C = \frac{1}{2} \times 5\% \times 1000 = 25$$
Similarly, since the YTM is expressed at an annual nominal rate, the period rate (or six month rate) is given by
$$r = \frac{1}{2} \times 6\% = 3\%$$
The value is therefore
\begin{align*}
P &= C \times a_{20\rceil 3\%} + \frac{F}{(1 + 3\%)^{20}} \\
&= 25 \times a_{20\rceil 3\%} + \frac{1000}{(1 + 3\%)^{20}} \\
&= 371.936872 + 553.675754 \\
&= 925.612626
\end{align*}
\end{example}

#### Bond Quotes

Bonds are normally quoted in the marketplace at a percentage of face value.  
Therefore, the market quote for the bond in the last example would be

$$\text{Quote} = \frac{925.612626}{1000} \times 100 = 92.561263$$

Convention to express in 6 decimal places.  
Even though the quote is expressed as a percentage of face value, we do not put a % symbol at the end!

### Premium and Discount Bonds

**Par Bond** - quoted price is equal to 100 (called the **par value**), and we say the bond is trading at a **par**.

**Discount Bond** - quoted price is less than 100, and we say the bond is trading at a **discount**.

**Premium Bond** - quoted price is more than 100, and we say the bond is trading at a **premium**.

Observation: a bond will trade at par exactly when the coupon rate _equals_ the yield to maturity.  
Consider a bond a with a face value $F$ paying coupons semi-annually with a yield to maturity $YTM$.  
Let $i = \frac{YTM}{2}$ be the period rate.  
Since the coupon rate equals the yield to maturity, we have the coupon amount is $C = F \times i$.  
Therefore

\begin{align*}
P &= C \times a_{n\rceil \frac{YTM}{2}} + \frac{F}{\left(1+\frac{YTM}{2}\right)^{n}} \\
&= Fi \times \left( \frac{1 - (1+i)^{-n}}{i} \right) + \frac{F}{(1+i)^{n}} \\
&= F
\end{align*}

#### Alternative Pricing Formula

Let $r$ be the coupon rate (per period) and $i$ the yield to maturity (per period), then using our original pricing formula and doing some substitutions and simplifications, we have

$$P = F + F(r-i)a_{n\rceil i}$$

From this expression, it is easy to see that a bond trades at a premium exactly when $r > i$ and a discount when $r < i$.

\begin{example}\label{example:YTMExampleAltEquation}
Consider a $5\%$ coupon Government of Canada 10 year bond with a YTM of $6\%$.  
Find the value of this bond using the alternative formula.

**Solution**  
The coupon is paid semi-annually, so the semi-annual rate is $r = \frac{5\%}{2} = 2.5\%$, and the semi-annual YTM is $i = \frac{6\%}{2} = 3\%$.  
A 10 year bond has $10 \times 2 = 20$ coupon periods, so therefore we have (based on 1000 face value)
\begin{align*}
P &= F + F(r-i)a_{n\rceil i}\\
&= 1000 + 1000(2.5\% - 3\%)a_{30\rceil 3\%} \\
&= 1000 - 74.387374 \\
&= 925.612626
\end{align*}
\end{example}

## Prices Between Coupon Rates

### Bond Prices

The formula for the price of a coupon bond only works if pricing the bond on a coupon date.

How do we find the price between coupon dates?  
The **Full Price** of a bond (also called the **dirty price** or **invoice price**) is the value of the bond when discounted back to the settlement day.

Let $f$ represent the fraction of a coupon period that "now" is before the next coupon,

$$f = \frac{\text{# of days from the last coupon payment to settlement date}}{\text{# days in coupon period}}$$

Ugly form: price of a coupon bond on a coupon date with $n$ periods remaining

$$FP_{\text{now}} = (1 \times i)^{f} \times \left( \frac{C}{(1+i)^{1}} + \frac{C}{(1+i)^{2}} + \cdots + \frac{C + F}{(1+i)^{n}} \right)$$

If we let $P_{0}$ be the price at time 0, we have the simpler formula for the full price of a bond,

$$FP_{\text{now}} = (1 \times i)^{f} \times P_{0}$$

\begin{example}\label{example:BondFullPrice}
Find the full price of a $5\%$ bond due June 30, 2022 with a YTM of $4\%$.  
Assume the settlement day is May 1, 2018 and that the bond pays semi-annual coupons.

**Solution**  
We begin by valuing the bond on the previous coupon date, which is June 30, 2017.  
Call that value $P_{0}$.  
How do we know that the coupon date was December 31, 2017?  
Since the bond pays coupons semi-annually and it matures on June 30, its coupon dates must be each June 30 and December 31 (six months later).

Note that there are 9 coupon periods from December 31, 2017 to June 30, 2022.  
Therefore, the price of the bond on December 31, 2017 is

\begin{align*}
P_{0} &= 25 \times a_{9\rceil \frac{4\%}{2}} + \frac{1000}{\left( 1 + \frac{4\%}{2} \right)^{9}} \\
&= 1040.81
\end{align*}

To get the full price on the settlement day we need to count the number of days from December 31, 2017 to May 18, 2018 (there are 138) as well as the total number of days in the coupon period from December 31, 2017 to June 30, 2018 (there are 181).

Therefore the fraction $f$ is $f = \frac{138}{181}$.  
Therefore, the full price today is the full price $P_{0}$ carried forward this fraction of a period, and

$$FP = (1 + 2\%)^{(138/181)} \times P_{0} = 1056.64$$
\end{example}


#### Clean Prices

Bonds are usually quoted in the market at their **clean price** which is the difference between the full price and the so called accrued interest.

The **accrued interest** is the amount of the next coupon payment that accrues to the seller.  
This is calculated on a simple interest basis, so

$$AI = f \times C$$

Then clean price of a bond (also called the **quoted price**) is as follows

$$CP = FP - AI$$

## Bond Amortization

### Bond Schedules

#### Amortization of Bond Premium and Discount

Recall

$$P = F + F(r - i)a_{n\rceil i}$$

So, the bond trades at a premium exactly when $r > i$ and a discount when $r < i$, and furthermore, the quantity

$$F(r - i)a_{n\rceil i}$$

is the size of the premium or discount.

As time passes, the number of periods until the bond matures, $n$, declines, hence the size of the premium or discount also declines.  
This represents what we call the **amortization of premium or discount**.  
So, over time, the size of the premium or discount is amortized away until it reaches zero at maturity of the bond.
