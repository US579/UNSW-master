## Chapter 10: Developing New Products and Services

**Product**

> A good, service, or idea consisting of a bundle of tangible and intangible attributes that satisfies consumers and is received in exchange for money or some other unit of value.

### The Variations of Products

**Product Line**

> A group of products that are closely related because they satisfy a class of need, are used together, are sold to same customer group, are distributed through the same type of outlets, and fall within a given price range.
>
> Each product line has its own marketing strategy.

**Product Mix**

> The number of product lines offered by a company.

Classifying products,

* type of user
    * consumer goods
    * business goods
* degree of tangibility

#### Classifying Consumer and Business Goods

**Consumer goods**

> Products purchased by the _ultimate consumer_.

**Business goods**

> Products that assist directly or indirectly in provided products for resale (also known as B2B goods, industrial goods, or organizational goods).

**Non-durable goods**

> An item consumed in or or a few uses.

**Durable goods**

> An item that usually lasts over an extended number of uses.

**Services**

> Activities, deeds, or other basic intangibles offered for sale to consumers to in exchange for money or something else of value.

##### Classification of Consumer Goods

**Convenience goods**

> Purchased frequently with minimum shopping effort, while consumers compare **shopping goods** on several attributes

**Specialty goods**

> Items for which the consumer makes a special effort to obtain

**Unsought goods**

> Products in which the consumer either does not know about or knows about but does not initially want

##### Classification of Business Goods

**Production goods**

> Become part of the final product

**Support goods**

> Purchased to assist in the production of the final product

Support goods includes:

* Installations

### What is a New Product

_New_ is difficult to define,

* newness compared with existing products
    * new if a product is functionally different from existing products
    * sometimes the newness is revolutionary and creates a whole new industry
    * sometimes additional features are added to an existing product to try to make it appeal to more customers
        * _feature bloat_, proliferation of extra features, can overwhelm consumers
* newness in legal terms
    * Industry Canada (regulates business practices) has determined that a product can be called "new" for only up to 12 months
* newness from the company's perspective
    * successful companies view newness and innovation at three levels
        * lowest level: least risk; a product line extension
        * middle level: significant jump in innovation or technology
        * highest level: true innovation, a truly revolutionary new product
* newness from the consumer's perspective

Consumption effects define newness, i.e. degree of new consumer learning needed, from low to high degree of new consumer learning required:

| Basis of Comparison | Continuous innovations | Dynamically continuous innovations | Discontinuous Innovations |
|---------------------|------------------------|------------------------------------|---------------------------|
| Definition | requires no new learning by consumers | disrupts consumer's normal routine but does not require totally new learning | requires new learning and consumption patterns by consumers |
| Examples | new improved shaver, detergent, and toothpaste | electric toothbrush, compact disc player, and automatic flash unit for cameras | wireless router, digital video recorder, and electric car |
| Marketing strategy | gain consumer awareness and wide distribution | advertise points of difference and benefits to consumers | educate consumers through product trial and personal selling |

### What Are New Services

Seven categories of new services, ranging from major innovations to simple style changes:

* Major service innovations
    * new core products for market that have not been previously defined
    * includes both new-service offerings and radical new processes
* Major process innovations
    * innovative new processes to deliver existing core services in new ways with greater value enhancement to the customer
* Service-line extensions
    * additions to existing lines of services
* Process-line extensions
    * less innovative than major process innovations
    * new ways of delivering existing services so that they offer greater convenience or a different experience to the customer
* Supplementary-service innovations
    * adding new elements to the core service or improving existing supplementary services that accompany the core service
* Basic service improvements
    * modest changes in the performance of the current service
* Style changes
    * focus on creating only incremental improvements to their current core offerings
    * there are the few service firms that focus on breakthrough or truly disruptive innovations can create entirely new markets or can dramatically reshape existing markets so that they can enjoy the benefits of unforeseen profits for a considerable length of time
        * e.g. Google

### Why New Products or Services Succeed or Fail

Marketing reasons for new-product failures

* insignificant "point of difference"
* incomplete market and product definition
    * lack of clear _protocol_
* too little market attractiveness
* poor execution of the marketing mix
* poor product quality on critical factors
* bad timing
* no economical access to buyers

**Protocol**

> A statement that, before product development beings, identifies (1) a well-defined target market; (2) specific customers' needs, wants, and preferences; and (3) what the product will be and do.


#### What were they thinking? Organizational problems new-product failure

* not really listening to the "voice of the consumer"
* skipping steps in the new-product process
* pushing a poorly conceived product into the market to generate quick revenue
* "Groupthink" in task force and committee meetings
* not learning critical takeaway lessons from past failures

#### Learning Review

What does "insignificant point of difference" mean as a reason for new-product failure?

> "Insignificant point of difference" means that a distinctive product feature may have been absent from the new-product.  
> A "point of difference" is necessary for a new product to defeat competitive ones -- through the possession of superior characteristics that deliver unique benefits to the user.

### The New-Product Process

> The stages a firm uses to identify business opportunities and convert them to a saleable good or service.

1. New-product strategy development
    * Purpose of stage
        * identify focus, structure, approach, and guidelines for innovation
    * Marketing information and methods used
        * Cross-functional teams
            * a small number of people from different departments in an organization who are mutually accountable to a common set set of performance goals
        * Six Sigma protocol
            * a means to "delight the customer" by achieving quality through a highly disciplined process that focuses on developing and delivering near-perfect products and services
            * success lies in determining what variables impact the results, measuring them, and making decisions based on data, not gut feeling
        * Lead users
            * a small group of potential product or service users who desire new products or services before the general market recognizes this need
2. Idea generation
    * Purpose of stage
        * develop a pool of concepts as candidates for new products
    * Marketing information and methods used
        * customer and supplier suggestions
            * lead users can play a vital role in providing suggestions or ideas
            * important to consider what non-customers may want; these customers usually a good reason for not being a customer
            * examining customer complaints
            * _crowdsourcing_: generating insights leading to actions based on massive numbers of people's ideas
        * employee and co-worker suggestions
            * everyone can contribute new product or service ideas
            * some companies "idea creation and idea management" training programs for employees
        * research and development breakthroughs
            * professional R&D and innovation labs outside the walls of large corporations are sources of _open innovation_ and can provide new-product ideas
        * competitive products
            * analyze the competition for new-product ideas
        * universities, inventors, and small technology firms
3. Screening and evaluation
    * Purpose of stage
        * separate good product ideas from bad ones inexpensively
        * internal and external evaluations of the new-product ideas to eliminate those that warrant no further effort
    * Marketing information and methods used
        * Internal approach
            * evaluates the technical feasibility of the proposal and whether the idea meets the objectives defined in the new-product strategy development step or stage
            * service-dominated offerings need to ensure that employees have the commitment and skills to meet customer expectations and sustain customer loyalty
        * External approach
            * Concept tests
                * external evaluations that consist of preliminary testing of the new-product idea
                * more useful with minor modifications of existing products
4. Business analysis
    * Purpose of stage
        * identify the product's feature and its marketing strategy, and make financial projections
        * last checkpoint before creating a _prototype_
    * Marketing information and methods used
        * product's key features
        * anticipated marketing mix strategy
        * economic, marketing, production, legal, and profitability analyses
5. Development
    * Purpose of stage
        * create the prototype product, and test it in the laboratory and on consumers
    * Marketing information and methods used
        * laboratory and consumer tests on product prototypes
        * for services, improving the delivery of customer service is critical; involves analyzing the entire sequence of steps or "service encounters" to improve the interactions between consumers and the service providers
6. Market testing
    * Purpose of stage
        * test product and marketing strategy in the marketplace on a limited scale
        * exposing actual products to prospective consumers under realistic purchase conditions to see if they will buy
    * Marketing information and methods used
        * standard test markets
            * a company develops a product and then attempts to sell it through normal distribution channels in a number of test-market cities
            * selected cities must be demographically representative of markets targeted for the new product
            * producer sells the product to distributors, wholesalers, and retailers
        * controlled test markets
            * contracting the entire test program to an outside service
        * simulated test markets
            * a technique that replicates a full-sale test market to a limited degree
        * ...when test markets do not work
            * some firms use _virtual reality test marketing_
7. Commercialization
    * Purpose of stage
        * position and offer product in the marketplace in full-scale and sales
        * most expensive stage for most new products
    * Marketing information and methods used
        * perceptual maps
        * product positioning
        * regional roll-outs
    * Slotting fee
        * a payment a manufacturer makes to place a new item on a retailer's shelf
    * Failure fee
        * a penalty payment a manufacturer makes to compensate a retailer for failed sales from its valuable shelf space
    * _Time to market_ (TtM) correlated to new-product success
        * i.e. needs to arrive on time
    * _Parallel development_
        * cross-functional team members, who conduct the simultaneous development of both the product and the product process, stay with the product from concept to production
    * _Fast prototyping_ in software development
        * "do it, try it, fix it" approach
        * encourages continuous improvements

#### Learning Review

What is the difference between internal and external screening and evaluation approaches used by a firm in the new-product process?

> Internal approach means the firm evaluates the technical feasibility of the proposal and whether the idea meets the objectives defined in the new-product strategy development step.  
> This is all done "in house".
>
> External evaluations of concept tests consist of preliminary testing of the new-product idea (rather than the actual product) with consumers.

What is a test market?

> Market testing is the exposing of actual products to prospective consumers under realistic purchase conditions to see if they will buy.

## Chapter 11: Managing Products and Brands

### Product Life Cycle

Stages of a product:

* Introduction stage
    * A product is first introduced to its intended market
    * Sales grow slowly and profit is minimal
        * Lack of profit is often the result of large investment costs in product development
    * Aggressive marketing and promotion expenditures to stimulate _primary demand_ (desire for product class)
    * Marketing objective: create consumer awareness and stimulate _trial_ -- the initial purchase of a product by a consumer
    * More competitors $\to$ creating _selective demand_ (demand for a specific brand)
    * High initial price may be used as part of a _skimming_ strategy to help the company recover the costs of development as well as capitalize on the price insensitivity of early buyers
    * To discourage competitive entry, price low, _penetration pricing_, and can help build unit volume, but must closely monitor costs
* Growth stage
    * Still marketing, although can be less aggressive, shifts to stimulating selective demand
    * Marketing pays attention to the following
        * Competition
        * Differentiating from competition
    * Increasing rate due to new people trying or using the product and a growing proportion of _repeat purchasers_
* Maturity stage
    * Slowing of total industry sales or product class revenue
    * Marginal competitors begin to leave the market
    * Marketing attention is often directed toward holding market share through further product differentiation and finding new buyers
    * A major consideration is to reduce overall marketing costs by improving promotional and distribution efficiency
* Decline stage
    * Frequently due to environmental changes; technological innovation often precedes the decline stage as newer technologies replace older technologies
    * Consumes a disproportionate share of management time and financial resources relative to their future future worth
    * _Deletion_
        * Dropping a product from a company's product line; most drastic
        * Affects the residual core of consumers
    * _Harvesting_
        * Retains the product but reduces marketing support costs
        * Purpose is to maintain the ability to meet customer requests

| Marketing Objective | Gain Awareness | Stress Differentiation | Maintain Brand Loyalty | Harvesting, Deletion |
|---|---|---|---|---|
| Competition | Few | More | Many | Reduced |
| Product | One | More versions | Full product line | Best sellers |
| Price | Skimming or penetration | Gain market share, deal | Defend market share, profit | Stay profitable |
| Place (Distribution) | Limited | More outlets | Maximum outlets | Fewer outlets |
| Promotion | Inform, educate | Stress points of difference | Reminder-oriented | Minimal promotion |

Four Dimensions of the Product Life Cycle:

* Length of the product life cycle
    * Consumer products have shorter life cycles than do business products
    * Availability of mass communication vehicles informs consumers faster and shortens life cycles
    * Rate of technological change tends to shorten product life cycles as new-product innovation replaces existing products
* Shape of the product life cycle
    * High-learning product
        * extended introductory period due to requirement of significant education of the customer
    * Low-learning product
        * sales begin immediately due to little learning required
        * benefits of purchase are readily understood
        * easily imitated by competitors
        * marketing strategy is to broaden distribution quickly, important to have the manufacturing capacity to meet demand
    * Fashion product
        * introduced, declines, and then seems to return
        * length of the cycles may be years or decades
    * Fad product
        * rapid sales on introduction and then an equally rapid decline
        * typically novelties and have a short life cycle
* The product level
    * _Product class_
        * the entire product category or industry
        * e.g., video game consoles and software
    * _Product form_
        * variations of a product within the product class
        * e.g., PSP, Wii; typically moves from introduction to maturity in 5 years
* The life cycle and consumers
    * Depends on sales to consumers, not all consumers rush to buy a product in the intro stage
    * Product diffuses, or spreads, through the population, called _diffusion of innovations_
    * **Five Categories and Profiles of Product Adopters**:
        * Innovators (2.5%)
            * venturesome; higher educated; use multiple information sources
        * Early adopters (13.5%)
            * leaders in social setting; slightly above average education
        * Early majority (34%)
            * deliberate; many informal social contacts
        * Late majority (34%)
            * skeptical; below average social status
        * Laggards (16%)
            * fear of debt; neighbors and friends are information sources
    * For any product to be successful, it must be purchased by innovators and early adopters; then moves onto early majority, late majority, and laggard once it has been accepted by previous two categories
    * **Factors Affecting the Adopting of a New Product in Introduction Stage**
        * usage barriers
            * product is not compatible with existing habits
        * value barriers
            * product provides no incentive to change
        * risk barriers
            * physical, economic, or social
        * psychological barriers
            * cultural differences or image
    * **Methods to Overcome Barriers**
        * warranties
        * money-back guarantees
        * extensive usage instructions
        * demonstrations
        * free samples to stimulate initial trial of new products (most popular mean to gain consumer trial)

### Managing the Product Life Cycle

* Role of a Product Manager
    * PM also called _brand manager_
    * Responsible for managing existing products through the stages of the life cycle, some also responsible for developing new products
    * Marketing responsibilities include developing and executing a marketing program for the product line described in an annual marketing plan and approving ad copy, media selection, and package design
    * Engage in extensive data analysis
    * Data supplemented with (1) category development index and (2) brand development index
        * indexes to help identify strong and weak market segments (usually demographic or geographic segments)
        * provide direction for marketing efforts
* Modifying the Product
    * _Product modification_
        * altering a product's characteristics, such as its quality, performance, appearance, features, or package, to try to increase and extend the product's sales
* Modifying the Market
    * Marking modification
        * strategy in which a company tries to find new customers, increase a product's use among existing customers, or create new-use situations
    * Finding new users
    * Increasing use
    * Creating new-use situation
* Repositioning the Product
    * _Product repositioning_
        * changing the place a product occupies in a consumer's mind relative to competing products
    * **Four Marketing Mix Elements**
        * Reacting to a competitor's position
            * A reason is a competitor's entrenched position is adversely affecting sales and market share
        * Reaching a new market
            * e.g., iced tea were poorly received in UK because British viewed it as leftover tea, i.e. not suitable for drinking, until the tea was carbonated and repositioned as a cold soft drink
        * Catching a rising trend
            * e.g., foods that offer health and dietary benefits like low fat
        * Changing the value offered
            * _Trading up_
                * adding value to a product (or line) through additional features or higher-quality materials
            * Trading down
                * reducing the number of features, quality, or price
            * Downsizing (make product/packaging smaller)
                * reducing the content of packages without changing package size and maintaining or increasing the package price
            * Trading down often exists when companies engage in downsizing; firms have been criticized for this practice

### Branding and Brand Management

Building and managing brands is critically important for marketers but also very difficult in a complex marketplace environment.

Environmental forces that impacts how marketers can effectively build and manage brands:

* empowered consumer
    * having the most critical impact on marketers and their branding strategies
* new media (e.g., social media and new technologies)
* increased pressure to demonstrate a return on investment
* importance of changing demographics
* ethics in business
* organizational change
* globalization

The goal is to create and deliver positive experiences with the brand in order to ensure customer satisfaction and strong customer loyalty to the brand.

* _Branding_
    * Activity in which an organization uses a name, phrase, design, symbols, or combination of these to identify its products and distinguish them from those of competitors
    * _Brand name_
        * any word, device (design, shape, sound, or colour), or combination of these used to distinguish a seller's goods or services
    * _Trade name_
        * a commercial, legal name under which a company does business
    * _Trademark_
        * identifies that a firm has legally registered its brand name or trade name so that the firm has its exclusive use
    * _Product counterfeiting_
        * low-cost copies of popular brands not manufactured by the original producers
        * a growing problem in Canada due to trademarks
* Brand Personality and Brand Equity
    * Brands offer more than product identification and a means to distinguish their products from competitors
    * _Brand personality_
        * a set of human characteristics associated with a brand name
        * marketers imbue a brand with a personality through advertising that depicts a certain user or usage situation and conveys certain emotions or feelings to be associated with the brand
    * _Brand equity_
        * the added value that a given brand name gives to a product beyond the functional benefits provided
        * provides the advantages of (1) competitive advantage and (2) consumers are often willing to pay a higher price for a product with brand equity
        * consumers willing to pay a premium for a brand over another when functional benefits are identical
    * Creating Brand Equity
        * (1) develop positive brand awareness and an association of the brand in consumers' mind with a product class or need to give the brand an identity
        * (2) marketer must establish a brand's meaning in the minds of consumers in two dimensions
            * performance-related dimension
            * abstract, imagery-related dimension
        * (3) elicit the proper consumer responses to a brand's identity and meaning
            * attention placed on how consumers think and feel about a brand
            * focuses on perceived quality, credibility, and superiority relative to other brands
            * feeling relates to the consumer's emotional reaction to a brand
        * (4) create a consumer-brand resonance evident in an intense, active loyalty relationship between consumers and the brand
            * most difficult step
            * deep psychological bond the resonance and the personal identification
        * forging an emotional connection leads to brand differentiation, customer loyalty, and evangelical promotion of the brand
    * Valuing Brand Equity
        * Provides financial advantage for brand owner
        * Brands can appreciate in value when marketed effectively; can also lose value when not managed properly
        * Difficult to attach monetary value to a brand
            * often use a variety of direct and indirect measures, including communication investment in the brand and customer loyalty to the brand
        * _Brand Licensing_
            * a contractual agreement whereby a company allows another firm to use its brand name, patent, trade secret, or other property for a royalty or fee
            

#### Picking a Good Brand Name

Five criteria for picking a good brand name:

1. The name should describe product benefits
2. The name should be memorable, distinctive, and positive
3. The name should fit with the company or product image
4. The name should have no legal restrictions
    * legal restrictions produce trademark infringement suits
    * regulatory restrictions can arise through improper use of words
5. The name should be simple

#### Branding Strategies

* _Multiproduct Branding_
    * used by a company of one name for all its product in a product class
    * sometimes called _family branding_ or _corporate branding_
    * makes possible _product line extensions_
    * risk with line extensions is that sales of an extension may come at the expense of other items in the product line
    * some companies employ _sub-branding_ (combines a corporate or family brand with a new brand)
    * a strong brand equity allows _brand extension_ (using a current brand name to enter a completely different product class)
        * too many uses for one brand name dilute the meaning of a brand for consumers
    * a variation on brand extension: _co-branding_
        * the pairing of two or more recognized brands on a single product or service
* _Multibranding_
    * a manufacturer's branding strategy giving each product a distinct name
    * higher promotional cost than multiproduct branding
    * company must generate awareness without benefits of any previous impressions
    * each brand name is unique to each market segment
        * failure in one does not affect other products
* _Private Branding_
    * when a company manufactures products but sells them under the brand name of a wholesaler or retailer
    * often called _private labelling_ or _reseller branding_
    * typically produces high profits for manufacturers and resellers
* _Cohort Brand Management_
    * the bundling of one company's multiple brands into a single marketing effort aimed at a common consumer group
    * typically done through online marketing
    * e.g. P&G's website is HomeMadeSimple.com
* _Mixed Branding_
    * a firm markets products under its own name and that of a reseller because the segment attracted by the reseller is different from its own market
* Emerging Branding Concepts
    * _social benefit brand_
        * brands that have a special added value in terms of their core environmental or social benefits provided to consumers
        * focuses include environmental sustainability (_green brands_) and a variety of non-ecological social benefits

### Packaging and Labeling

**Packaging**

> any container in which a product is offered for sale and on which label information is communicated

**Label**

> an integral part of the package that typically identifies the product or brand

Creating customer value through packaging and labeling:

* Communication benefits
    * include directions on how to use the product and the composition of the product, required by law
    * also have seals and symbols, either government-required or commercial seals of approvals
* Functional benefits
    * storage, convenience, protection, or product quality
* Perceptual benefits
    * package and label shape, colour, and graphics distinguish one brand from another, convey a brand's positioning, and build brand equity
    * enhance brand recognition and facilitate the formation of strong, favourable, and unique brand associations

Contemporary packaging and labeling challenges:

* Connecting with customers
    * challenges in creating aesthetic and functional design features that attract attention and deliver value in use
* Environmental concerns
    * concern about the growth of solid waste and shortage of viable landfill sites
    * leads to attention in the amount, composition, and disposal of packaging material
* Health, safety and security issues
    * many consumers believe packages are safe and secure, regardless of cost
* Cost reduction
    * must find innovative ways to cut packaging costs while delivering value to their customers due to material cost increase

### Product Warranty

**Warranty**

> a statement indicating the liability of the manufacturer for product deficiencies

* Express warranties
    * written statements of liabilities
    * government requires greater disclosure on express warranties to indicate limited-coverage or full-coverage alternative
* Limited-coverage warranty
    * states the bounds of coverage and areas of non-coverage
* Full warranty
    * no limits of non-coverage
* Implied warranties
    * responsibilities for product deficiencies to the manufacturer

Warranties affect product evaluation.  
Limited warranty products tend to receive less positive evaluations.  
Also hold a significant marketing advantage.

Warranty important in light of increasing product liability claims.  
Early part of the 20th century, courts protected companies, but the trend now is _strict liability_ rulings, where a manufacturer is liable for any product defect, whether it followed reasonable research standards or not.  
Issue contested by companies and consumer advocates.

## Chapter 12: Managing Services

### The Service Economy

Services

> Intangible activities, benefits, or satisfactions that an organization provides to consumers in exchange for money or something else of value.

Responsible for over 70% of Canada's total economic output, as well as over 75% of its workforce.  
Vancouver hot service businesses: life coaching, personal training, health care for the elderly, massage therapy, interior design, immigration consulting, per services, and beauty salons.

#### The Uniqueness of Services

_Pure good_ implies that the consumer obtains benefits from the good alone without any added value from service.  
_Pure service_ assumes there is no "goods" element to the service that the customer receives.  
Most services contain some goods element, and most goods offer some service.

The 4 I's of services

* Intangibility
    * A major marketing need is to make them tangible or to show the benefits of using a service
* Inconsistency
    * Inconsistent in quality of service, so challenges in developing, pricing, promoting, and delivering services
* Inseparability
    * Hard to separate the production and consumption
    * Hard to separate the person delivering the service from the brand itself
        * bad employee $\to$ poor perception of the service received by the consumer
* Inventory
    * _Idle production capacity_
        * when the service provider is available but there is no demand
    * Inventory cost of a service: cost of paying the person used to provide the service, along with any needed equipment

Cost of inventory (from low to high):

* Real estate agency; hair salon
* Insurance company
* Dry cleaner
* Auto repair centre
* Restaurant
* Hotel
* Amusement park
* Airline; hospital

#### Service Continuum

The 4 Is of services differentiate services from goods in most cases, but most products sold cannot be defined as pure goods or pure services.

**Service continuum**

> a range from the tangible to the intangible or goods-dominant to service-dominant offerings available in the marketplace

$$\text{Service-dominated offering (intangible)} \longleftarrow~\text{Balanced}~\longrightarrow \text{Product-dominated offering (tangible)}$$

Good to distinguish between _core_ services from _supplementary_ services.  
Supplementary services often allow service providers to differentiate their offer from competitor, and may add value for consumers.

#### How Consumers Purchase Services

* Purchasing a Service
    * More difficult for consumers to evaluate services before purchase
    * 3 aspects of a service
        * Search Properties - Determined before purchase (e.g., price)
        * Experience Properties - Experience properties may only be discerned during or after consumption (e.g., restaurant food quality)
        * Credence Properties - Characteristics that the consumer may find impossible to evaluate even after purchase and consumption
    * Marketers turn to personal sources of information (e.g., early adopters, opinion leaders) to ensure customer satisfaction in order to achieve positive word-of-mouth referral

#### Customer Contact Audit

> a flow chart of the points of interaction between consumer and service provider

Points of interaction also referred as _contact points_, _touchpoints_, _service encounter elements_. 

Important in high-contact services.

#### Post-purchase Evaluation

Evaluation by comparing expectations about service offering with the actual experience.  
Expectations are influenced by word-of-mouth communications, personal needs, past experience, and marketing communication activities.  
Actual experience determined by the way an organization delivers the service.

* Gap analysis
    * expectation vs reality
* SERVQUAL
    * a way to conduct gap analysis
    * judges service quality
    * 5-10% of unhappy service customers offer direct feedback
    * multidimensional research instrument designed to measure service quality by capturing respondents' expectations and perceptions along the 5 dimensions of service quality
        * Reliability (5 questions)
            * ability to perform the promised service dependably and accurately
        * Assurance (4 questions)
            * respectful, considerate personnel who listen to customers and answer their questions
        * Tangibles (4 questions)
            * appearance of physical facilities, equipment, personnel, and communications materials
        * Empathy (5 questions)
            * knowing the customer and understanding their needs
            * approachable and available
        * Responsiveness (4 questions)
            * willingness to help customers and provide prompt service
    * 22 expectation items and 22 perceptions items

#### Managing the Marketing of Services: The 7 P's

* Product (Service)
    * product concept with features and benefits desired by customers
    * branding is an important aspect of product concept

Knowing your customer and reaching them in a creative and integrated manner

* _Promotional Mix_
    * combination of one or more communication tools: advertising, personal selling, sales promotion, public relations, and direct marketing
    * used to (1) inform prospective buyers about the benefits of the product, (2) persuade them to try it, and (3) remind them later about the benefits they enjoyed by using the product

    * concept of designing marketing communications program that coordinate all promotional activities -- advertising, personal selling, sales promotion, public relations, and direct marketing -- to provide a consistent message across all audiences and to maximize the promotional budget and impact of the communications

### The Communication Process

* _Communication_
    * The process of conveying a message to others and requires a **source**, a **message**, a **channel of communication**, a **receiver**, and the **processes of encoding and decoding**
    * Source - is a sender -- such as a company or salesperson -- who has information to convey
    * Message - is the information to be transmitted, such as the details of a new weight-reduction drink
    * Channel of communication - the means of conveying a message to a receiver
    * Receivers - are consumers who read, hear, or see the message
* Encoding and Decoding
    * Essential to communications, but very difficult, could be lost in transformation or travel
    * Encoding - process of having the sender transform an abstract idea into a set of symbols
    * Decoding - process of having the receiver take a set of symbols, the message, and transform them back to an abstract idea
    * Field of experience - similar understanding and knowledge; mutually shared field of experience needed for two sides for effective communication
* Feedback
    * Response - the impact the message had on the receiver's knowledge, attitudes, or behaviour
    * Feedback - the communication flow from receiver back to sender that helps the sender know whether the message was decoded and understood as intended
* Noise
    * Noise - extraneous factors that can work against effective communication by distorting a message or the feedback received
    * e.g., a misprint in a newspaper ad or salesperson's message is misunderstood by a prospective buyer

### The Promotional Elements

* _Advertising_
    * any paid form of non-personal communication about an organization, good, service, or idea by an identified sponsor
    * no immediate feedback loop
    * has control on what, whom, and when by company
* _Personal selling_
    * two-way flow of communication between a buyer and seller designed to influence a person's or group's purchase decision
    * immediate feedback, controls whom and coverage by salesperson
* _Public relations_
    * a form of communication management that seeks to influence the feelings, opinions, or beliefs held by customers, prospective customers, shareholders, suppliers, employees, and other publics about a company and its product or services
    * _publicity_
        * a non-personal, indirectly paid presentation of an organization, good, or service
        * lack of the user's control over it
* _Sales promotion_
    * a short-term inducement of value offered to arouse interest in buying a good or service
    * lose effectively if conducted continuously
* _Direct marketing_
    * uses direct communication with consumers to generate a response in the form of an order, a request for further information, or a visit

* Identifying the Target Audience
    * Easier to develop a promotion program with more knowledge about its target audience's profile
* Specifying Promotion Objectives
    * **Hierarchy of effects** - describes the consumer's sequence of stages a consumer goes through in making a buying decision, from initial awareness to eventual action
        * Awareness - consumer's ability to recognize and remember the product or brand name
        * Interest - an increase in the consumer's desire to learn about some of the features
        * Evaluation - consumer's appraisal of the product or brand on important attributes
        * Trial - consumer's actual first purchase and use
        * Adoption - through a favourable experience leads to repeated purchase and use
    * Should posses 3 qualities: (1) be well designed for a well-defined target audience, (2) be measurable, and (3) cover a specified time period
* Setting the Promotion Budget
    * _Percentage of sales_
        * allocating funds to advertising as a percentage of past or anticipated sales, in terms of either dollars or units sold
        * simple and provides a financial safeguard
        * major fallacy that implies sales cause promotion
    * _Competitive parity_
        * matching the competitors' absolute level of spending or the proportion per point of market share
        * also referred as _matching competitors_ or _share of market_
        * should not be only determinant in setting budget: competition may have very different promotional objectives $\to$ different level of promotion expenditures
    * _All-you-can-afford_
        * allocating funds to promotion only after all other budget items are covered
        * fiscally conservative
        * used as if no knowledge of promotion-to-sales relationship or what its promotional objectives are
    * _Objective and task_
        * the company (1) determines its promotion objectives, (2) outlines the tasks to accomplish these objectives, and (3) determines the promotion cost of performing these tasks
        * strength of other budgeting methods integrated into here
        * challenge lies in judgment required to determine the required tasks
* Selecting the Right Promotional Tools
    * The large number of possible combination of the promotional tools means many combinations can achieve the same objective
    * Analytical approach and experience important in this step
    * Important to assess the relative importance of the various tools
* Designing the Promotion
    * Successful designs are often the result of insight regarding the consumers' interests and purchasing behaviour
* Scheduling the Promotion
    * Importance in scheduling most effective timing of use
    * Promotion schedule outlines order and frequency of each promotional tool during the campaign
        * can be influenced by factors such as seasonality and competitive promotion activity

### Executing and Evaluating the IMC Program

IMC audit: a tool used to evaluate the company's current process.  
Analyzes the internal communication network of the company;
identifies key audiences;
evaluates customer databases;
assess messages in recent ads, public relations releases, packaging, video news releases, signage, sales promotion programs, direct mail, and online, including presence on social media;
and determine the IMC expertise of the company and agency personnel.

Companies must monitor consumer content, respond to inconsistent messages, and answer questions from individual customers.

Pre-test each design before it is actually used to allow for changes and modifications that improve its effectiveness.

Post-tests are recommended to evaluate the impact of each promotion and the contribution of the promotion toward achieving the program objectives.
